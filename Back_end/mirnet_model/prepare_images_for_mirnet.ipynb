```python
from PIL import Image
import matplotlib.image
```

```python
from pathlib import Path

def write_a_series_of_images_to_pngs(series_of_images, fname_base):
	for i in range (0, series_of_images.shape[0]):
		print (f"i = {i}")
		#take image and normalize it
		tmp = series_of_images[i, :, :, 0].numpy()
		tmp_min = np.min(tmp)
		tmp_max = np.max(tmp)
		tmp = (tmp-tmp_min)/(tmp_max - tmp_min)
		#repeat 3 times to make it into an input for png writer
		b = np.repeat(tmp[:, :, np.newaxis], 3, axis=2)
		#print (f"b.shape = {b.shape}")
		#print (f"b = \n{b}")
		Path(fname_base).mkdir(parents=True, exist_ok=True)
		matplotlib.image.imsave(f'{fname_base}/{i}.png',  b)



    
```

```python

write_a_series_of_images_to_pngs (data_train_X_series_of_images , "./our485/low" )
write_a_series_of_images_to_pngs (data_train_Y_series_of_images , "./our485/high") 
write_a_series_of_images_to_pngs (data_test_X_series_of_images  , "./eval15/low" )  
write_a_series_of_images_to_pngs (data_test_Y_series_of_images  , "./eval15/high")
```
